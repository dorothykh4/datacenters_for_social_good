```python
import pandas as pd

# Load the dataset (skip the first two descriptive rows)
df = pd.read_csv("average_price_of_electricity.csv", skiprows=2)

# Rename columns so they are easier to work with
df.columns = [
    "state",
    "res_2025", "res_2024",
    "com_2025", "com_2024",
    "ind_2025", "ind_2024",
    "transport_2025", "transport_2024",
    "all_2025", "all_2024"
]

# Keep only the relevant columns
df = df[["state", "ind_2025"]]

# Rename industrial column to price
df = df.rename(columns={"ind_2025": "electricity_price"})
```

```python
print(df.isna().sum())
```

```python
print(df[df["electricity_price"].isna()])
```

```python
# Make the price column numeric (non-numeric values to Nan)
df["electricity_price"] = pd.to_numeric(df["electricity_price"], errors="coerce")

# Remove rows that are NaN in the price column 
# (it is only Census Division\nand State and Technical notes; no states are removed)
df = df.dropna(subset=["electricity_price"])

# Reset index
df = df.reset_index(drop=True)

# Save cleaned dataset
df.to_csv("electricity_price_clean.csv", index=False)
```
