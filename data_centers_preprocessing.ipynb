```python
import pandas as pd

# Import dataset
df = pd.read_csv("im3_open_source_data_center_atlas_v2026.02.09.csv")

# Keep relevant columns for the research question
df = df[[
    "id",
    "state",
    "state_abb",
    "county",
    "county_id",
    "lat",
    "lon",
    "type",
    "sqft"
]]
```

```python
duplicates = df[df.duplicated(subset='id', keep=False)]
print(duplicates)
df = df.drop_duplicates(subset='id', keep='first')
```

```python
# Check missing values in each column
print("Missing values per column:")
print(df.isna().sum())
```

```python
# Check labels in the 'type' column
print("\nUnique values in 'type':")
print(df["type"].unique())

# Count how many of each type exist
print("\nCounts for each type:")
print(df["type"].value_counts())

# Final dataset preview
print("\nCleaned dataset preview:")
print(df.head())
```
